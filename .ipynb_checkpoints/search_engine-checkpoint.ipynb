```python
#streaming version of the code so I don't use too much memory
#loading the data line by line rather then the whole thing at once
import gzip 

with gzip.open("data/goodreads_books.json.gz", 'r') as f: 
    line = f.readline()
#reading line by line
```

```python
#use json module to load the json using loads (load string)
#creates python dict
import json 
json.loads(line)
```

```python
#format the lines how we want
def parse_fields(line): 
    data = json.loads(line)
    return{
        "book_id": data["book_id"],
        "title": data["title_without_series"],
        "ratings": data["ratings_count"],
        "url": data["url"],
        "cover_image": data["image_url"]
    }
```

## Adding Books to list that contain <10 ratings

```python
#parse each line and read each line to add
books_titles = []
with gzip.open("data/goodreads_books.json.gz", 'r') as f: 
    while True: 
        line = f.readline()
        if not line: 
            break
        fields = parse_fields(line)

        try: 
            ratings = int(fields["ratings"]) #int for ratings, if valueerror, continue
        except ValueError: 
            continue
        if ratings > 10: #only adding books that contain ratings more than 10 ratings
            books_titles.append(fields)
```

```python
import pandas as pd

titles = pd.DataFrame.from_dict(books_titles)
print(titles)
```

```python
#make ratings numerical
titles["ratings"] = pd.to_numeric(titles["ratings"])
```

```python
#regex to make titles more uniform
titles["modified_title"] = titles["title"].str.replace("[^a-zA-Z0-9 ]","", regex = True)
```

```python
titles
```

```python
#lowercase and removed white space
titles["modified_title"] = titles["modified_title"].str.lower()
titles["modified_title"] = titles["modified_title"].str.replace("\s+"," ", regex=True)
```

```python
titles
```

```python
titles = titles[titles["modified_title"].str.len() >0 ]
```

```python
#write to json file 
titles.to_json("books_titles.json")
```

## TF-IDF to Cosine Similarity Search Function
Next step is to use inverse document frequency to put all the book titles into matrix form so the search engine can use that format to search for book titles. This is why we cleaned the title names to the 'modifed_title' column. (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

tfidf = vectorizer.fit_transform(titles["modified_title"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import re

def make_clickable(val): 
    return '<a target="_blank" href="{}">Goodreads</a>'.format(val)

def show_image(val): 
    return '<img src="{}" width=50></img>'.format(val)

def search(query , vectorizer):
    processed = re.sub("[^a-zA-Z0-9 ]", "", query.lower())
    query_vec = vectorizer.transform([processed])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -10)[-10:] #find the indices of similarity
    results = titles.iloc[indices]
    results = results.sort_values("ratings", ascending=False)
    return results.head(5).style.format({'url':  make_clickable, 'cover_image': show_image})
```

```python
search("Everything I never told you", vectorizer)
```

## Building an Interactive Recommendation Engine from our Search Engine base

Using widgets in Jupyter Notebooks


```python
# Import necessary libraries for search and recommendations
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import re
import ipywidgets as widgets
from IPython.display import display

# Mock data for titles and tf-idf matrix (replace with actual data)
titles = None  # Replace with a DataFrame containing book titles, ratings, URLs, and cover images
tfidf = None  # Replace with the TF-IDF matrix for the book titles

# Define the helper functions for clickable links and images
def make_clickable(val): 
    return f'<a target="_blank" href="{val}">Goodreads</a>'

def show_image(val): 
    return f'<img src="{val}" width=50></img>'

# Search function (from user-provided code)
def search(query, vectorizer):
    processed = re.sub("[^a-zA-Z0-9 ]", "", query.lower())
    query_vec = vectorizer.transform([processed])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -10)[-10:]  # Find the indices of the top matches
    results = titles.iloc[indices]
    results = results.sort_values("ratings", ascending=False)
    return results.head(5).style.format({'url': make_clickable, 'cover_image': show_image})

# Recommendation logic (from uploaded notebook)
def recommend_books(book_id, similarity_matrix, book_mapping):
    # Get the index of the book_id
    book_index = book_mapping.get(book_id)
    if book_index is None:
        return ["Book not found in the database."]
    # Find similar books based on the similarity matrix
    similar_indices = np.argsort(similarity_matrix[book_index])[::-1][1:6]
    similar_books = [titles.iloc[idx] for idx in similar_indices]
    return similar_books

# Widgets
input_box = widgets.Text(description='Book Title:', placeholder='Enter a book title...', layout=widgets.Layout(width='50%'))
output_box = widgets.Output()
search_button = widgets.Button(description='Search', button_style='success', tooltip='Click to get recommendations', icon='search')

# Callback function for the widget
def on_search_click(b):
    book_title = input_box.value.strip()
    if not book_title:
        with output_box:
            output_box.clear_output()
            print("Please enter a valid book title.")
        return

    # Search for the book
    with output_box:
        output_box.clear_output()
        if tfidf is None or titles is None:
            print("Mock data used. Replace `titles` and `tfidf` with actual data for real results.")
        else:
            results = search(book_title, vectorizer=tfidf)
            print("Search Results:")
            display(results)

# Link button to the callback
search_button.on_click(on_search_click)

# Display the widget
app_layout = widgets.VBox([input_box, search_button, output_box])
display(app_layout)
```
